import boto3

#create table 

def create_table(dynamodb=None):
    if not dynamodb:
        dynamodb = boto3.resource('dynamodb', endpoint_url="http://localhost:8000")

    table = dynamodb.create_table(
        TableName='Etoken',
        KeySchema=[
            {
                'AttributeName': 'user_id',
                'KeyType': 'HASH'  # Partition key
            },
            {
                'AttributeName': 'password',
                'KeyType': 'RANGE'  # Sort key
            }
        ],
        AttributeDefinitions=[
            {
                'AttributeName': 'user_id',
                'AttributeType': 'S'
            },
            {
                'AttributeName': 'password',
                'AttributeType': 'S'
            },

        ],
        ProvisionedThroughput={
            'ReadCapacityUnits': 1,
            'WriteCapacityUnits': 1
        }
    )
    return table


if __name__ == '__main__':
    etoken_table = create_table()
    print("Table status:", etoken_table.table_status)

import boto3

def get_item(dynamodb=None):
    if not dynamodb:
        dynamodb = boto3.resource('dynamodb', endpoint_url="http://localhost:8000")
    table = dynamodb.Table('Etoken')
    response = table.get_item(
        Key={
            'user_id': 'amar',
            'password': 'test123'
        }
    )
    item = response['Item']
    print(item)

if __name__ == '__main__':
    get_item()

```python
#fetch all records from the table

import boto3

def get_all_items(dynamodb=None):
    if not dynamodb:
        dynamodb = boto3.resource('dynamodb', endpoint_url="http://localhost:8000")
    table = dynamodb.Table('Etoken')
    response = table.scan()
    data = response['Items']
    return data

if __name__ =='__main__':
    all_data = get_all_items()
    print(all_data)
```
